# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
#Chose a sate date and end date of your trip.
# November 1st through November 10th
```

```python
# Determine max date
max_date = session.query(Measurement.date). \
        order_by(Measurement.date.desc()). \
        first(). \
        date
max_date
```

```python
# Calculate the date 1 year ago from the last data point in the database
last_twelve_months = dt.date(2017, 8, 23) - dt.timedelta(days=365)
last_twelve_months
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

preceipitation_data = session.query(Measurement.date, Measurement.tobs). \
                        filter( Measurement.date >= last_twelve_months).all()

preceipitation_data
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
# Sort the datadrame by date 

precipitation_data_df = pd.DataFrame(preceipitation_data, columns=['Date', 'Precipitation'])
precipitation_data_df.set_index('Date', inplace=True)
precipitation_data_df.head(10)
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
precipitation_data_df.plot(title=" Last 12 Months of Precipitation")
plt.savefig("1st Chart")
plt.show
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
precipitation_data_df.describe()
```

```python
# Design a query to show how many stations are available in this dataset?
stations = session.query(Station).count()
stations
```

```python
# What are the most active stations? List the stations and the counts in decending order. 
stations = session.query(Measurement.station,func.count(Measurement.station)). \
    group_by(Measurement.station). \
    order_by(func.count(Measurement.station).desc()). \
    all()
stations
```

```python
# Most active station
active_station = stations[0][0]
active_station
```

```python
# Calcualte the lowest, highest, and average temperature of the most active station

stats = session.query(func.max(Measurement.tobs), func.min(Measurement.tobs), func.avg(Measurement.tobs)). \
                filter(Measurement.station == active_station).all()
stats
```

```python
# Choose the station with the highest number of temperature observations 
# Query the last 12 months of temperature observation data for this station 

temp_observations = session.query(Measurement.date, Measurement.tobs). \
                    filter( Measurement.station == active_station). \
                    filter( Measurement.date >= last_twelve_months).all()
 
temp_observations
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
# Sort the datadrame by date 
temp_observations_df = pd.DataFrame(temp_observations, columns=['Date', 'Temp'])
temp_observations_df.set_index('Date', inplace=True)
temp_observations_df.head(10)
```

```python
# Plot the results as a histogram
temp_observations_df.plot(kind='hist', bins=12)
plt.title("Temperature Observations for Station USC00519281", fontsize=20)
plt.savefig("Hist Chart")
plt.show
```
